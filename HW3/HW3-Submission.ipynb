Xét:

$2Me^{-2\times\epsilon^2N} \leq b$

$\Leftrightarrow N \geq \frac{ln(\frac{b}{2M})}{-2\times\epsilon^2}$

Xây dựng function bên dưới:


```python
import math
def N_min(b,M,epsilon):
    return math.log(b/(2*M))/(-2*epsilon**2)
```

## Câu 5 (1 điểm)

Theo bài giảng, ta có:
- Nếu $H$ không có break point thì growth function bằng $2^N$.
- Nếu $H$ có break point $k$ thì growth function bị chặn trên bởi một đa thức theo $N$ bậc $k-1$  cố định. 

Lần lượt xét từng trường hợp:

i) $f_{i}(N) = 1 + N $

- $f_{i}(N) < 2^N$ tại N = 2 
- $f_{i}(N)$ là một hàm đa thức bậc 1


Do đó, $f_{i}(N)$ có thể là một growth function


ii) $f_{ii}(N) = 1 + N + C^i_N = \frac{N^2}{2}+\frac{N}{2}+1$. 

- $f_{ii}(N) < 2^N$ tại N = 3
- $f_{ii}(N)$ là một hàm đa thức bậc 2


Do đó, $f_{ii}(N)$ có thể là một growth function

iii) $f_{iii}$ = $\sum_{i=1}^{[\sqrt{N}]}C^i_N$ .
- $f_{iii}(N) < 2^N$ tại N = 1
- $f_{iii}$ không phải là hàm đa thức, đây làm hàm mũ. 

Do đó, $f_{iii}(N)$ có thể là một growth function

iv) $f_{iv}$ = $2^{[\frac{N}{2}]}$

- $f_{iv}(N) < 2^N$ tại N = 1
- $f_{iv}$ không phải là hàm đa thức, đây làm hàm mũ. 

Do đó, $f_{iv}(N)$ có thể là một growth function

v) $f_{v}$ = $2^{N}$

- $f_{v}(N) = 2^N$ với mọi N (Hiển nhiên đúng)

Do đó, $f_{v}(N)$ có thể là một growth function

Vậy chỉ có trường hợp i, ii, v có thể là một growth function

Chọn B
